```python
import pandas as pd
import matplotlib.pyplot as plt

# Definindo os dados em um dicionário
data = {'Provincia': ['Cabo Delgado', 'Gaza', 'Inhambane', 'Manica', 'Maputo Cidade', 'Maputo Provincia', 'Nampula', 'Niassa', 'Sofala', 'Tete', 'Zambezia'],
        'Populacao afetada': [545, 271, 102, 327, 793, 765, 620, 79, 538, 244, 789]}

# Criando um objeto DataFrame do pandas com os dados
df = pd.DataFrame(data)

# Gerando o gráfico de barras
df.plot(kind='bar', x='Provincia', y='Populacao afetada', figsize = (12,6))

# Personalizando o gráfico
plt.title('População afetada pelo coronavírus em Moçambique em 2020')
plt.xlabel('Provincia')
plt.ylabel('População afetada')
plt.xticks(rotation=90)

# exportando o gráfico
plt.savefig('grafico_moz_covid_2020')

# Exibindo o gráfico
plt.show()
```
